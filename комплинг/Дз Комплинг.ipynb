```python
with open('роб.txt', 'r') as f:
        text = f.read()
text=str(text)
text=text.replace("\n"," ")
text
```

```python
##Разбиваем предложения регулярным выражением, которое будет делить предложения по **.?!**
```

```python
sentences = re.split(r"[.!?]", text)
sentences = [sent.strip(" ") for sent in sentences]
```

```python
sentences
```

```python
from pprint import pprint
from nltk.tokenize.punkt import PunktSentenceTokenizer, PunktTrainer
```

```python
trainer = PunktTrainer()
trainer.INCLUDE_ALL_COLLOCS = True
trainer.train('\n'.join(text))
 
tokenizer = PunktSentenceTokenizer(trainer.get_params())
```

```python
tokenizer._params.abbrev_types.add('15')
```

```python
print(tokenizer._params.abbrev_types)
```

```python
[re.sub('[\n\t]', ' ', x)[-10:] for x in tokenizer.tokenize(sentences[5])[:15]]
```

```python
gold = tokenizer.tokenize(text)
```

```python
gold
```

```python
tp = 0
fp = 0
fn = 0

for sent in gold:
    if len(sent_tokenize(sent)) == 1:
        tp += 1
    else:
        fp += 1

for i in range(len(gold)-1):
    sent1, sent2 = gold[i], gold[i+1]
    sent = ' '.join([sent1, sent2])
    if len(sent_tokenize(sent)) == 2:
        tp += 1
    else:
        fn += 1

precision = (tp/(tp+fp))
recall = (tp/(tp+fn))
f1 = 2*(precision*recall)/(precision+recall)
print('Precision - ', precision)
print('Recall - ', recall)
print('F1 - ', f1)
```
